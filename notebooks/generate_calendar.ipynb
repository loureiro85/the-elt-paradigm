```python
import pandas as pd


def generate_calendar_table(start='2000-01-01', end='2050-12-31'):
    """Generate calendar table dimension."""
    WEEKDAY = {
        0: 'sunday'
    }

    df = pd.DataFrame({'date': pd.date_range(start, end)})
    df['date_id'] = df['date'].dt.strftime('%Y%m%d').astype(int)


    df['year'] = df['date'].dt.year
    df['month'] = df['date'].dt.month
    df['month_name'] = df['date'].dt.month_name()
    df['weekday'] = df['date'].dt.day_of_week
    df['weekday_name'] = df['date'].dt.day_name()
    df['month_day'] = df['date'].dt.day
    df['week'] = df['date'].dt.isocalendar().week
    df['quarter'] = df['date'].dt.quarter
    df['semester'] = (df['quarter'] + 1) // 2
    df['year_month'] = df['date'].dt.to_period('M')

    df.set_index('date_id', inplace=True)
    return df


df = generate_calendar_table()
df.to_csv('data/raw/calendar.csv')
df
```
